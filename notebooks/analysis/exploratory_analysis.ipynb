### Section 2

```python
s2 = df.filter(regex='s2')

tcc = df[df['s1_orientacion_teo']=='Terapias Cognitivas/Comportamentales']
psa = df[df['s1_orientacion_teo']=='Psicoanálisis']
eclectico = df[df['s1_orientacion_teo']=='Ecléctico (más de una de estas opciones)']
sistemico = df[df['s1_orientacion_teo'] =='Sistémica']


s2_tcc = tcc[['s2_evidencia_cientifica','s2_experiencia_personal','s2_entrenamiento_clinica',
                      's2_tratamiento_preferencia_consultantes','s2_intuicion','s2_terapia_personal']].mean()

s2_psa =  psa[['s2_evidencia_cientifica','s2_experiencia_personal','s2_entrenamiento_clinica',
                      's2_tratamiento_preferencia_consultantes','s2_intuicion','s2_terapia_personal']].mean()

s2_eclectico = eclectico[['s2_evidencia_cientifica','s2_experiencia_personal','s2_entrenamiento_clinica',
                      's2_tratamiento_preferencia_consultantes','s2_intuicion','s2_terapia_personal']].mean()

s2_sistemico = sistemico[['s2_evidencia_cientifica','s2_experiencia_personal','s2_entrenamiento_clinica',
                      's2_tratamiento_preferencia_consultantes','s2_intuicion','s2_terapia_personal']].mean()

df2 = {
    'PSA': s2_psa.round(2),
    'TCC': s2_tcc.round(2),
    'ECLECTICO': s2_eclectico.round(2),
    'SISTEMICA': s2_sistemico.round(2)
}

df2 = pd.DataFrame(df2)
df2
```

```python
ax = df2.plot(kind='barh', figsize=(10,10), width=0.8)

ax.set_xlabel('Puntaje')
ax.set_ylabel('Variables')
ax.set_title('Medias segun marco teo')
plt.show()
```
